# Calculation: FOV and resolution for 3 setups

### Mitutoyo NA 0.5, x50

```python
lamda = 580 # nm
spot_size_px = 1500  # px
px_size_um = 6.5 # um / px
magnification = 50
NA = 0.5

total_power = 30 # uW

image_size_um = spot_size_px * px_size_um
fov = image_size_um / magnification
print(f'FOV = {fov} µm')

# Lateral resolution: d = 0.61 λ / NA
d = 0.61 * lamda/ NA
print(f'Lateral resolution = {d} nm')

# Axial resolution dz = 2λ / NA^2
dz = 2*lamda/(NA**2)
print(f'Axial resolution = {dz} nm')

power_density = total_power / (fov**2)
print('Power density: %0.2e uW/um2' %power_density)
print('Power density: %0.2e W/cm2' %(power_density/100))
```

### Nikon NA 0.95, x60

```python
lamda = 580 # nm
spot_size_px = 1500  # px
px_size_um = 6.5 # um / px
magnification = 60
NA = 0.95

total_power = 30 # uW

image_size_um = spot_size_px * px_size_um
fov = image_size_um / magnification
print(f'FOV = {fov} µm')

# Lateral resolution: d = 0.61 λ / NA
d = 0.61 * lamda/ NA
print(f'Lateral resolution = {round(d, 1)} nm')

# Axial resolution dz = 2λ / NA^2
dz = 2*lamda/(NA**2)
print(f'Axial resolution = {dz} nm')

power_density = total_power / (fov**2)
print('Power density: %0.2e uW/um2' %power_density)
print('Power density: %0.2e W/cm2' %(power_density/100))
```

## Fiber collimation

$$f = \frac{D \pi \omega}{4 \lambda}$$

Where f : focal length of the lens

$\omega$ : mode fiber diameter

$\lambda$: wavelength

```python
D = 1130e-6
omega = 3.2e-6
lamda = 515e-9
f = D*np.pi*omega/(4*lamda)
print(f'Focal length f = {round(f/1, 4)} m')
```
